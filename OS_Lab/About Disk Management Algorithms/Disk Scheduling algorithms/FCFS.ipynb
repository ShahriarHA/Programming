```python
departed_queue_list = []
departed_queue_list.append(r_W_head)
print(len(request_queue_list))
```

```python
if len(request_queue_list) == 0:
    print("There is no request which contains track no!")
else:
    current_r_W_head = r_W_head
    total_no_track_movement = 0
    flag = 0
    
    for i in range(len(request_queue_list)):
        #difference between the current head and the request
        print(i)
        seek_time = abs(request_queue_list[i] - current_r_W_head)
        
        current_r_W_head = request_queue_list[i]
        total_no_track_movement += seek_time
        
        departed_queue_list.append(request_queue_list[i])
        flag+=1
        
    print(flag)
    for j in range(flag):
        print(j)
        request_queue_list.pop()
print("requst queue : ", request_queue_list)
print("departed queue : ", departed_queue_list)
print("total seek time: ",total_no_track_movement)
        
#         request_queue_list.pop(i)

        
#     print("queue have")
```
